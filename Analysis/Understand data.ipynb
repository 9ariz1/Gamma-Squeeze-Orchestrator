```python
import yfinance as yf
import pandas as pd

# Stock symbol
stock_symbol = "AAPL"

# Download stock data
data = yf.download(stock_symbol, start="2024-01-01", end="2024-06-01")

data.columns = data.columns.get_level_values(0)

print("\n===== FIRST 5 ROWS =====")
print(data.head())

print("\n===== LAST 5 ROWS =====")
print(data.tail())

print("\n===== DATA INFO =====")
print(data.info())

print("\n===== STATISTICAL SUMMARY =====")
print(data.describe())
```

```python
import yfinance as yf
import pandas as pd

stock_symbol = "TSLA"   # try TSLA / NVDA / GME
ticker = yf.Ticker(stock_symbol)

expiry = ticker.options[0]
option_chain = ticker.option_chain(expiry)
calls = option_chain.calls

calls['volume'] = calls['volume'].fillna(0)

current_price = ticker.history(period="1d")['Close'].iloc[-1]
print("Current Stock Price:", current_price)

atm_calls = calls[
    (calls['strike'] >= current_price - 50) &
    (calls['strike'] <= current_price + 50)
]

top_oi = atm_calls.sort_values(by='openInterest', ascending=False).head(10)
top_volume = atm_calls.sort_values(by='volume', ascending=False).head(10)

gamma_candidates = pd.concat([top_oi, top_volume]).drop_duplicates()

print("\n===== POTENTIAL GAMMA ZONE (TOP ACTIVITY) =====")
print(gamma_candidates[['strike', 'volume', 'openInterest']])
```

```python
important_cols = ['strike', 'volume', 'openInterest', 'impliedVolatility']

print("\n===== IMPORTANT CALL DATA =====")
# Replace NaN volume with 0
calls['volume'] = calls['volume'].fillna(0)
print(calls[important_cols].head())

```
